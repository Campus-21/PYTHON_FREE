```python
print ("Hello World!")
print ("Hello", "World!") #функция с несколькими аргументами

print('End test', end='#')   # Конец строки. По усолчанию  /n
print('Sep', 'test', 'end', sep = '*') # Разделитель между аргументами.  


# 3 метода форматирования строк
#========================================
# % - метод

print('Hello, %s!' % 'World')
      
    
#'%d', '%i', '%u'	Десятичное число.
#'%o'	Число в восьмеричной системе счисления.
#'%x'	Число в шестнадцатеричной системе счисления (буквы в нижнем регистре).
#'%X'	Число в шестнадцатеричной системе счисления (буквы в верхнем регистре).
#'%e'	Число с плавающей точкой с экспонентой (экспонента в нижнем регистре).
#'%E'	Число с плавающей точкой с экспонентой (экспонента в верхнем регистре).
#'%f', '%F'	Число с плавающей точкой (обычный формат).
#'%g'	Число с плавающей точкой. с экспонентой (экспонента в нижнем регистре), если она меньше, чем -4 или точности, иначе обычный формат.
#'%G'	Число с плавающей точкой. с экспонентой (экспонента в верхнем регистре), если она меньше, чем -4 или точности, иначе обычный формат.
#'%c'	Символ (строка из одного символа или число - код символа).
#'%r'	Строка (литерал python).
#'%s'	Строка (как обычно воспринимается пользователем).
#'%%'	Знак '%'.


#=========================================

# Вывод экранированных последовательностей.  Escape-последовательности.

print('\newline') 
#Если после символа "\" сразу нажать клавишу Enter то это позволит продолжать запись с новой строки.
print('\\')
#Позволяет записать символ обратного слеша.
print('\"') 
#Позволяет записать один символ апострофа.
print('\"') #	Позволяет записать один символ кавычки.
print('\a') #	Гудок встроенного в систему динамика.
print('\b') #	Backspace, он же возврат, он же "пробел назад" – удаляет один символ перед курсором.
print('\f') #	Разрыв страницы.
print('\n') #	Перенос строки (новая строка).
print('\r') #	Возврат курсора в начало строки.
print('\t') #	Горизонтальный отступ слева от начала строки (горизонтальная табуляция).
print('\v') #	Вертикальный отступ сверху (вертикальная табуляция).
print('\x05') # 	Шестнадцатеричный код символа (две шестнадцатеричные цифры xhh).
print('\ooo') #	Восьмеричный код символа (три восьмеричные цифры ooo).
print('\0') #	Символ Null.
#print('\uhhhh') #	Шестнадцатеричный код 16-битного символа Юникода (символ кодируемый двумя байтами).
#print('\Uhhhhhhhh') #	Шестнадцатеричный код 32-битного символа Юникода (символ кодируемый четырьмя байтами).
#print('\other')
#Под other понимается любая другая последовательность символов. 
#Не является экранированной последовательностью (остается без изменений с сохранением в строке символа "\"). 

#=========================================

# Форматирование строк. Отступы

# #	Значение будет использовать альтернативную форму.
printf("%x", 13)
printf("%#x", 13)
printf("%X", 13)

#0"	Свободное место будет заполнено нулями.
print("%04d", 15)

#-"	Свободное место будет заполнено пробелами справа.
# "	Свободное место будет заполнено пробелами справа.
#+"	Свободное место будет заполнено пробелами слева.






```
